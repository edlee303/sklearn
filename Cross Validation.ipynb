```python
from sklearn.model_selection import GridSearchCV
from sklearn import datasets, svm, tree, neighbors
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
def asses_svm(data, target):
    param_grid = [
      {'C': [1, 10, 100, 1000], 'kernel': ['linear']},
      {'C': [1, 10, 100, 1000], 'gamma': [0.001, 0.0001], 'kernel': ['rbf']}
     ]
    
    training_set = [data, target]
    svc = svm.SVC()
    clf = GridSearchCV(svc, param_grid)
    clf.fit(*training_set)
    
    return pd.DataFrame(clf.cv_results_).sort_values('mean_test_score', ascending=False).head()
```

```python
def asses_tree(data, target):
    param_grid = {
        'criterion':['gini','entropy'],
        'splitter':['best','random'],
        'max_depth':list(range(1,10)),
        'min_samples_split':list(range(2,4))}
    
    training_set = [data, target]
    dtc = tree.DecisionTreeClassifier()
    clf = GridSearchCV(dtc, param_grid)
    clf.fit(*training_set)
    
    return pd.DataFrame(clf.cv_results_).sort_values('mean_test_score', ascending=False).head()
```

```python
def asses_neighbors(data, target):
    param_grid = {
        'n_neighbors':list(range(1,10)),
        'weights':['uniform','distance'],
        'algorithm':['ball_tree','kd_tree','brute']}
    
    training_set = [data, target]
    knc = neighbors.KNeighborsClassifier()
    clf = GridSearchCV(knc, param_grid)
    clf.fit(*training_set)
    
    return pd.DataFrame(clf.cv_results_).sort_values('mean_test_score', ascending=False).head()
```

```python
iris = datasets.load_iris()
data = iris.data
target = iris.target
asses_tree(data, target)
```

```python
hp = datasets.load_boston()
prices = pd.DataFrame(hp.target, columns=['price'])
prices['quantile'] = pd.qcut(prices['price'], 4, labels=False)
target = prices['quantile']
data = hp.data
```

```python
asses_svm(data, target)
```
